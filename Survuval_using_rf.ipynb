```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')
data.info()
data.dtypes
```

```python
data.head()
```

```python
#segregating null values in the dataset
nul = pd.concat([data.isnull().sum()] , axis = 'columns')
nul
```

Data Cleaning - replacing null values with mean or mode depending on nature of data

```python
data["Age"] = data["Age"].fillna(data["Age"].mean())
```

```python
data["Embarked"] = data["Embarked"].fillna(data["Embarked"].mode()[0])
data["Cabin"] = data["Cabin"].fillna(data["Embarked"].mode()[0])
```

```python
data["Pclass"] = data["Pclass"].apply(str)
```

```python
data.dtypes
```

```python
for i in data.dtypes[data.dtypes == 'object'].index :
  dummy = data.pop(i)
  data = pd.concat([data , pd.get_dummies(dummy , prefix = i)], axis = 1)
```

```python
data.head()
```

```python
data.dtypes
```

```python
y = data.pop('Survived')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
x_train, x_test, y_train, y_test = train_test_split(data, y, test_size = 0.20)
```

```python
rf = RandomForestClassifier(25)
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import roc_curve, auc
false_positive, true_positive, t = roc_curve(y_test, y_pred)
roc_auc = auc(false_positive, true_positive)
```
